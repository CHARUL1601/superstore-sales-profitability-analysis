## **DATA ANALYSIS ON SUPERSTORE SALES DATA**

Objective: Working on superstore sales dataset to figure out profit - discount analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## DATA LOADING AND EXPLORATION | CLEANING

```python
#converting the csv file to dataframe

sales_data = pd.read_csv("/content/Superstore.csv")
```

```python
#looking into top 5 rows of the data to understand the structure
sales_data.head(5)
```

```python
sales_data.columns
```

```python
sales_data = sales_data.set_index("Row ID")
```

```python
sales_data.head()
```

```python
print(f"The Number of Rows are {sales_data.shape[0]}, and columns are {sales_data.shape[1]}.")
```

```python
sales_data.info()
```

```python
sales_data.isnull().sum()
```

```python
#filling up the missing Postal Code values based on Region
sales_data["Postal Code"] = sales_data["Postal Code"].fillna(sales_data.groupby("Region")["Postal Code"].transform(lambda x: x.mode()[0] if not x.mode().empty else np.nan))
```

```python
sales_data.isnull().sum()
```

```python
sales_data["Postal Code"]
```

```python
# postal code is in float, coverting this values into int
sales_data["Postal Code"] = sales_data["Postal Code"].astype(int)
```

```python
# converting dates to datetime fromat
sales_data["Order Date"] = pd.to_datetime(sales_data["Order Date"])
sales_data["Ship Date"] = pd.to_datetime(sales_data["Ship Date"])

sales_data[["Order Date", "Ship Date"]].dtypes
```

```python
# creating columns for year and month for further analysis
sales_data["Year"] = sales_data["Order Date"].dt.year
sales_data["Month"] = sales_data["Order Date"].dt.to_period("M")
```

Findings
1. Initially there were 20 columns which has now become 22 since Year and Month are added
2. There were null values in the column Postal Code which has been filled with codes according to region and its type has been converted to int
3.  Order Date and Ship Data data types has changed to datetime

**Milestone 1** : Data is free from missing values

## TRENDS RELATED TO ORDERS ON A YEARLY, MONTHLY AND DAILY BASIS

```python
#finding total number of orders per day
unique_date = sales_data.groupby(["Order Date","Year", "Month"])["Order ID"].nunique().reset_index()
unique_date = unique_date.rename(columns = {"Order ID" : "No. of Orders"})
unique_date
```

```python
# finding maximumn and mean order per day
print(f"Maximum orders in a day were {unique_date["No. of Orders"].max()}")
print(f"Mean orders per day were {unique_date["No. of Orders"].mean() :.2f}")
```

```python
yearly_orders_summary = unique_date.groupby("Year")["No. of Orders"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_orders_summary, x="Year", y="No. of Orders", marker="o", color="royalblue")
plt.title("Total Annual Orders Show Consistent Year-Over-Year Increase", fontsize=16)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(yearly_orders_summary["Year"].unique())
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

```python
monthly_orders_analysis = unique_date.groupby(["Year", "Month"])["No. of Orders"].sum().reset_index()

# Convert the 'Month' Period object to an integer month number
monthly_orders_analysis["Month"] = monthly_orders_analysis["Month"].dt.month

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_orders_analysis, x="Month", y="No. of Orders", hue="Year", marker="o", palette="viridis")
plt.title("Monthly Orders Show Consistent Year-End Peaks and Early-Year Dips Across All Years", fontsize=16)
plt.ylabel("Total Number of Orders", fontsize=12)
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.legend(title="Year")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

```python
monthly_avg_orders = monthly_orders_analysis.groupby("Month")["No. of Orders"].mean().reset_index()
monthly_avg_orders["Month_Name"] = monthly_avg_orders["Month"].map({1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"})

lowest_month = monthly_avg_orders.loc[monthly_avg_orders["No. of Orders"].idxmin()]
highest_month = monthly_avg_orders.loc[monthly_avg_orders["No. of Orders"].idxmax()]

monthly_increase_percent = ((highest_month["No. of Orders"] - lowest_month["No. of Orders"]) / lowest_month["No. of Orders"]) * 100

print("--- Monthly Order Analysis ---")
print(f"Overall Average Orders per Month: {monthly_orders_analysis["No. of Orders"].mean():.2f}")
print(f"Lowest Performing Month: {lowest_month["Month_Name"]} with an average of {lowest_month["No. of Orders"]:.2f} orders.")
print(f"Highest Performing Month: {highest_month["Month_Name"]} with an average of {highest_month["No. of Orders"]:.2f} orders.")
print(f"Orders increase by {monthly_increase_percent:.2f}% from the lowest month to the highest month on average.")

# --- Yearly Order Analysis Metrics ---
print("\n--- Yearly Order Analysis ---")
print(f"Total orders in {yearly_orders_summary.loc[0, "Year"]}: {yearly_orders_summary.loc[0, "No. of Orders"]}")
for i in range(1, len(yearly_orders_summary)):
    current_year = yearly_orders_summary.loc[i, "Year"]
    previous_year = yearly_orders_summary.loc[i-1, "Year"]
    current_year_orders = yearly_orders_summary.loc[i, "No. of Orders"]
    previous_year_orders = yearly_orders_summary.loc[i-1, "No. of Orders"]
    year_on_year_growth = ((current_year_orders - previous_year_orders) / previous_year_orders) * 100
    print(f"Total orders in {current_year}: {current_year_orders} (a {year_on_year_growth:.2f}% growth compared to {previous_year})")
```

```python
day_names_list = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
unique_date["DayOfWeek"] = unique_date["Order Date"].dt.dayofweek
unique_date["DayName"] = unique_date["DayOfWeek"].map(lambda x: day_names_list[x])

daily_yearly_orders_analysis_line = unique_date.groupby(["Year", "DayName", "DayOfWeek"])["No. of Orders"].sum().reset_index()

# Sort by DayOfWeek to ensure correct order on x-axis for plotting
daily_yearly_orders_analysis_line = daily_yearly_orders_analysis_line.sort_values(by="DayOfWeek")

g = sns.relplot(
    data=daily_yearly_orders_analysis_line,
    x="DayName", y="No. of Orders",
    col="Year", col_wrap=2,  # Create a separate subplot for each year, wrap after 2 columns
    kind="line", marker="o", height=4, aspect=1.5
)
g.set_axis_labels("Day of Week", "Total Number of Orders")
g.set_titles("Year: {col_name}")
g.set_xticklabels(rotation=45)

plt.suptitle("Daily Order Trends Across Years Show Varying Weekly Patterns", y=1.02, fontsize=16)

plt.tight_layout()
plt.show()
```

```python
overall_daily_orders = daily_yearly_orders_analysis_line.groupby(["DayName", "DayOfWeek"])["No. of Orders"].sum().reset_index()
overall_daily_orders = overall_daily_orders.sort_values(by="DayOfWeek")

lowest_day_overall = overall_daily_orders.loc[overall_daily_orders["No. of Orders"].idxmin()]
highest_day_overall = overall_daily_orders.loc[overall_daily_orders["No. of Orders"].idxmax()]

print(f"Overall total orders by day of week:")
for index, row in overall_daily_orders.iterrows():
    print(f"- {row["DayName"]}: {row["No. of Orders"]} orders")

print(f"\nThe day with the lowest total orders across all years is {lowest_day_overall["DayName"]} with {lowest_day_overall["No. of Orders"]} orders.")
print(f"The day with the highest total orders across all years is {highest_day_overall["DayName"]} with {highest_day_overall["No. of Orders"]} orders.")
```

**Milestone 3:**
1. Total number of orders have significantly increased from 2015 to 2018
2. Highest perofrming month is Novemebr with an average of 188.25 orders which is increase of 381.45% as compared to the lowest performing month February
3. No significant pattern found between No. of Orders and Days of Week. It is similar for 2015,2016 which varies from the pattern found for 2017 and 2018



## TRENDS RELATED TO SALES ON A YEARLY, MONTHLY AND DAILY BASIS

```python
total_sales = sales_data.groupby(["Order Date","Year", "Month"])["Sales"].sum().reset_index()
total_sales
```

```python
print(f"Maximum sales in a day were {total_sales["Sales"].max():.2f}")
print(f"Mean sales per day were {total_sales["Sales"].mean():.2f}")
```

```python
yearly_sales_summary = total_sales.groupby("Year")["Sales"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_sales_summary, x="Year", y="Sales", marker="o", color="darkgreen")
plt.title("Total Annual Sales Show Consistent Year-Over-Year Increase", fontsize=16)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(yearly_sales_summary["Year"].unique())
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

```python
monthly_sales_analysis = total_sales.groupby(["Year", "Month"])["Sales"].sum().reset_index()

# Convert the 'Month' Period object to an integer month number for plotting
monthly_sales_analysis["Month_Num"] = monthly_sales_analysis["Month"].dt.month

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales_analysis, x="Month_Num", y="Sales", hue="Year", marker="o", palette="viridis")
plt.title("Monthly Sales Trends Showing Consistent Year End Peaks", fontsize=16)
plt.ylabel("Total Sales", fontsize=12)
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.legend(title="Year")
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

plt.show()
```

```python
monthly_avg_sales = monthly_sales_analysis.groupby("Month_Num")["Sales"].mean().reset_index()
monthly_avg_sales["Month_Name"] = monthly_avg_sales["Month_Num"].map({1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"})

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_avg_sales, x="Month_Name", y="Sales", marker="o", color="darkorange")
plt.title("Average Monthly Sales Across All Years Showing Peaks in March, September and November", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Sales", fontsize=12)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.show()
```

```python
monthly_sales_pivot = monthly_sales_analysis.pivot_table(index="Year", columns="Month_Num", values="Sales")

plt.figure(figsize=(14, 6))
sns.heatmap(monthly_sales_pivot, cmap="YlGnBu", fmt=".2f", linewidths=0.5, linecolor="black")
plt.title("Monthly Sales Heatmap Across Years summarizing Increase in Sales in Each year Along with Seasonal Peaks", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.xticks(ticks=np.arange(12) + 0.5, labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.yticks(rotation=0)
plt.show()
```

```python

lowest_sales_month = monthly_avg_sales.loc[monthly_avg_sales["Sales"].idxmin()]
highest_sales_month = monthly_avg_sales.loc[monthly_avg_sales["Sales"].idxmax()]

monthly_sales_increase_percent = ((highest_sales_month["Sales"] - lowest_sales_month["Sales"]) / lowest_sales_month["Sales"]) * 100

print("--- Monthly Sales Analysis ---")
print(f"Overall Average Sales per Month: {monthly_sales_analysis["Sales"].mean():.2f}")
print(f"Lowest Performing Month: {lowest_sales_month["Month_Name"]} with an average of ${lowest_sales_month["Sales"]:.2f} in sales.")
print(f"Highest Performing Month: {highest_sales_month["Month_Name"]} with an average of ${highest_sales_month["Sales"]:.2f} in sales.")
print(f"Sales increase by {monthly_sales_increase_percent:.2f}% from the lowest month to the highest month on average.")

print("\n--- Yearly Sales Analysis ---")
print(f"Total sales in {yearly_sales_summary.loc[0, "Year"]}: ${yearly_sales_summary.loc[0, "Sales"]:.2f}")
for i in range(1, len(yearly_sales_summary)):
    current_year = yearly_sales_summary.loc[i, "Year"]
    previous_year = yearly_sales_summary.loc[i-1, "Year"]
    current_year_sales = yearly_sales_summary.loc[i, "Sales"]
    previous_year_sales = yearly_sales_summary.loc[i-1, "Sales"]
    year_on_year_sales_growth = ((current_year_sales - previous_year_sales) / previous_year_sales) * 100
    print(f"Total sales in {current_year}: ${current_year_sales:.2f} (a {year_on_year_sales_growth:.2f}% growth compared to {previous_year})")
```

**Milestone 4:**
1. Sales have significantly increased from 2015 to 2018
2. Highest perofrming month is Novemebr and lowest performing month is February

# PROFIT AND DISCOUNT ANALYSIS

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales_data, x="Discount", y="Profit", alpha=0.6)
plt.title("Increase in Discount leads to a Decrease in Profit", fontsize=16)
plt.xlabel("Discount", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.7)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.show()
```

```python
average_profit_by_discount = sales_data.groupby("Discount")["Profit"].mean().reset_index()
print("Average Profit by Discount Level:")
print(average_profit_by_discount)
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(data=average_profit_by_discount, x="Discount", y="Profit", color="steelblue")
plt.title("Higher Discount Levels has Lower Average Profits", fontsize=16)
plt.xlabel("Discount Percentage", fontsize=12)
plt.ylabel("Average Profit", fontsize=12)
plt.xticks(rotation=45)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)
plt.show()
```

```python
correlation = sales_data["Profit"].corr(sales_data["Discount"])
print(f"Pearson correlation coefficient between Profit and Discount: {correlation:.2f}")
```

**Milestone 5:**

1. A scatter plot visualization revealed a tendency for profits to decrease, often leading to losses, as the 'Discount' percentage increases.
2. Quantitative analysis showed that various discount levels, specifically 0.30, 0.40, 0.50, 0.60, 0.70, and 0.80, resulted in negative average profits.
3. A bar chart visually confirmed that higher discount percentages are generally associated with lower average profits.
4. The Pearson correlation coefficient of -0.22 indicates a weak inverse linear relationship between 'Profit' and 'Discount', meaning as discounts increase, profit tends to slightly decrease.
